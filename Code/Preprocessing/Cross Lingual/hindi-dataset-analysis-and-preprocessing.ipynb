```python




# Combine both DataFrames
combined_df = pd.concat([sarcastic_df, non_sarcastic_df], ignore_index=True)

combined_df = combined_df.drop_duplicates()

# Randomize the data
df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save to a new CSV file if needed


print("Combined and randomized data:")
print(df.head())
```

```python
import pandas as pd
from googletrans import Translator

# Initialize the translator
translator = Translator()


# Function to translate text
def translate_text(text):
    try:
        translation = translator.translate(text, src='auto', dest='en')
        return translation.text
    except Exception as e:
        return text

# Translate the "Text" column
df['preprocessed_text'] = df['text'].apply(translate_text)


# Save the modified DataFrame back to a CSV file
df.to_csv('translated_english.csv', index=False)
```
